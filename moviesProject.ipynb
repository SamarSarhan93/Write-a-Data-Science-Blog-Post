## Data visualization

### Count of each Genre Type

```python
plt.title('Count of each Genre Type')
sns.set(style="darkgrid")
ax = sns.countplot(y="Genre", data=df, palette="Set1")
```

### Count of each Year

```python
plt.title('Count of each Year')
sns.set(style="darkgrid")
ax = sns.countplot(y="Year", data=df, palette="Set1")
```

### The Genre has the longest runtime minutes

```python
sns.barplot(data = df , x = "timeMin" , y='Genre', palette="Set1" )
plt.title('The Genre has the longest runtime minutes')
plt.xlabel('timeMin')
```

### Run time Minutes Density

```python
plt.title('Run time Minutes Density')
sns.set(style="darkgrid")
ax = sns.kdeplot(data=df['timeMin'], palette="Set1")
```

### The start Year for each title Type

```python
sns.violinplot(data = df , x = 'Year', y = 'Genre', palette="Set1")
plt.title('The start Year for each title Type')
```
